```python
import os
import json
import pandas as pd

from tqdm.notebook import tqdm
tqdm.pandas()

DATA_DIR = "../../data/challenges/Ferrara2023-citizens-individualChallenges.json"
OUT_DIR = "../../out/parquet/"
```

```python
with open(os.path.join(DATA_DIR)) as f:
    data = [json.loads(line) for line in f]

display(len(data))
data = [c for c in data if c['concept']['modelName'] not in ['survey']]
display(len(data))
data = [pd.json_normalize(c, sep='.') for c in tqdm(data)]
```

```python
df_individual = pd.concat([df.dropna(axis=1, how='all') for df in data], ignore_index=True)
print(df_individual.columns)
df_individual
```

```python
df_individual['startTime'] = pd.to_datetime(df_individual['concept.start'], cache=True).dt.date
df_individual['target'] = df_individual['concept.fields.target'].astype(int)
df_individual['periodTarget'] = df_individual['concept.fields.periodTarget'].apply(lambda x: 0 if pd.isna(x) else x).astype(int)
df_individual['counterName'] = df_individual['concept.fields.counterName'].apply(lambda x: x.replace(' ', '_'))
df_individual['state'] = df_individual['concept.state']
df_individual = df_individual[[
    'playerId',
    'startTime',
    'counterName',
    'target',
    'periodTarget',
    'state']]
df_completed = df_individual[df_individual['state'].isin(['COMPLETED', 'FAILED'])]
display(df_completed['state'].unique())
df_completed.to_parquet(os.path.join(OUT_DIR, 'individual-challenge.parquet'), compression='gzip')
```
